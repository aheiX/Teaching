```python
import math
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
def get_data(func_type):
  data = {
      'x': [],
      'y_type': [], 
      'y_para': [],
      'y_value': []
  }

  x_values = [i/10 for i in range(1, 101)]

  a = 1
  for b in range(5, 16):
    real_b = b/10

    data['x'].extend(x_values)
    data['y_type'].extend([func_type for i in x_values])
    data['y_para'].extend(['a=' + str(a) + ', b=' + str(real_b) for i in x_values])

    if func_type == 'Exponentialfunktion':
      data['y_value'].extend([a * math.exp(real_b*i) for i in x_values])

    elif func_type == 'Potenzfunktion':
      data['y_value'].extend([a*(i**real_b) for i in x_values])

    elif func_type == 'Logarithmusfunktion':
      data['y_value'].extend([a + (real_b * np.log(i)) for i in x_values])
    
    elif func_type == 'Hyperbolisch':
      data['y_value'].extend([a + (real_b /i) for i in x_values])

  return data

for func_type in ['Exponentialfunktion', 'Potenzfunktion', 
                  'Logarithmusfunktion', 'Hyperbolisch']:
  df = pd.DataFrame(data=get_data(func_type))
  fig = px.line(df, x='x', y='y_value',
                color='y_para',
                template='simple_white', 
                title=func_type,
                width=800, height=500)
  fig.show()
```
